# Task 3: Exploratory Data Analysis - Retail

[Exploratory data analysis (EDA) is used by data scientists to analyze and investigate data sets and summarize their main characteristics, often employing data visualization methods.
EDA is primarily used to see what data can reveal beyond the formal modeling or hypothesis testing task and provides a provides a better understanding of data set variables and the relationships between them. It can also help determine if the statistical techniques you are considering for data analysis are appropriate. Originally developed by American mathematician John Tukey in the 1970s, EDA techniques continue to be a widely used method in the data discovery process today.]

```python
#visualizing the corelation of features
f,ax = plt.subplots(figsize =(20,15))
sns.heatmap(df.corr(method='pearson'), annot=True, cmap="inferno", lw =6, linecolor='black')
```

```python
#observing region
sns.pairplot(df, hue='Region')
```

```python
#observing category
sns.pairplot(df,hue='Category')
```

```python
sns.countplot(x = df['Category'])
```

```python
sns.countplot(x = df['Region'])
```

Observation 1 :
1. In the Ship Mode , Standard Class has the highest count meaning it is of higher priorty to the customers.

2. In the Segment, Consumer has the highest count meaning they use more of the shippings.

3. In the Category, Office Supplies Category has the highest count.

4. In the Region, West region has the highest customers'count.

```python
plt.figure(figsize=(10,10))
df['Sub-Category'].value_counts().plot.pie(autopct= '%3.1f%%')
plt.show()
```

Observation 2 :

The top subcategories are Binders, Papers, Furnishings, Phones, Storage, Art, Accessories and Chairs while the bottom subcategories are Copiers and Machines.

```python
#observing the top selling states
top_20 = df.groupby('State')['Quantity'].sum().reset_index().sort_values(by = 'Quantity', ascending = False)
sns.catplot(x = 'State', y = 'Quantity', data = top_20.head(20), kind = 'bar', aspect = 3, palette='icefire')
plt.title('Top 20 States', fontsize = 20)
plt.xticks(rotation = 50, ha = 'right', fontsize = 20)
plt.yticks(fontsize = 15)
plt.xlabel('States', fontsize = 15)
plt.ylabel('Quantities ordered', fontsize = 15)
```

Observation 3 :

We observe that the top selling states are California, New York and Texas.

```python
# Profit vs Quantity
sns.lineplot(x='Quantity',y='Profit',label='Profit',data=df, color="red")
plt.legend()
plt.show()
```

```python
df.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['blue','red'],figsize=(10,6))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

Observation 4 :
    
Profit and sales are maximum in the consumer segment and minimum in the home office segment.

```python
df.groupby('Region')[['Profit','Sales']].sum().plot.bar(color=['yellow','green'],figsize=(10,6))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

Observation 5 :
    
Profit and Sales are maximum in the West Region and minimum in the South Region

```python
plt.figure(figsize=(15,10))
plt.title('Segment wise Sales in Each Region')
sns.barplot(x='Region',y='Sales',data=df,hue='Segment',order=df['Region'].value_counts().index,palette='rocket_r')
plt.xlabel('Region',fontsize=12)
plt.show()
```

Observation 6:
    
Segment - Wise Sales are almost same in every region.

```python
ps = df.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['blue','red'],figsize=(12,10))
plt.title('Profit/loss and  Sales across states')
plt.xlabel('States')
plt.ylabel('Profit/loss and Sales')
plt.show()
```

Observation 7:

1. Countries having high profits include California, New York
2. Countries having loss include Texas, Pennysylvania, Illinois, Ohio.

```python
df.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['red','cyan'],figsize=(10,7))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

Observation 8 :
    
Technology and Office Supplies have the highest profit wheras furniture has least profit.

```python
ps = df.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['blue','red'],figsize=(12,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('Sub-Category')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

Observation 9 :
    
1. Phones have the highest sales and Copiers have the highest profit.
2. Chairs have less profit than sub-category phone, although their sales are almost same.
3. Tables and Book-Cases are facing loss.

Observation 10:
    
The weak areas where one can make more profit is:
1. South Region is the least shipped area. We can improve by using marketing strategies like advertisements as well as discounts to attract customer attention and increase the sales in that particular region.
2. The least preferred shipping modes are Same Day and First Class modes. So, we can indulge in more advertising and additional benefits in order to attract more customers.
3. Since furnitures have less profit , more advertising and discount strategies will be needed for more public attention
